## Graph from two lists

```python
x = list(range(-10,11)) # Generating a list using range()
y = [n**2 for n in x] # Generating a list using a List Comprehension

print(x)
print(y)
```

```python
#help(plt.plot)
```

```python
plt.plot(x, y, 'purple')
plt.show()
```

### Adding a description to axes

```python
plt.plot(x, y, 'blue')

plt.title('Function y = x^2 da -10 a +10') 

plt.xlabel('x') 
plt.ylabel('y') 

plt.show()
```

## Two curves on the same graph

Available colours https://matplotlib.org/examples/color/named_colors.html

```python
# Generating points with NumPy
x = np.arange(0, np.pi*2, 0.1) 
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin,'slateblue', label="sin(x)")
plt.plot(x, y_cos,'teal', label="cos(x)")

# Describing the graph
plt.title('Trigonometry: sin(x) e cos(x)')
plt.xlabel('x')

# Creating a legend
plt.legend() 

# Showing the grid
plt.grid(True, which='both') 

plt.show()
```

# SciPy and distributions

List of distributions in `scipy.stats`   
https://docs.scipy.org/doc/scipy/reference/stats.html#continuous-distributions

We will show an example of a **Normal distribution**
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm

```python
import numpy as np
import scipy.stats as stats
```

```python
normal = stats.norm

y = normal.rvs(10, scale=3, size=1000)
x = np.linspace(0, y.max(), 1000)

# Generating the curve
param = normal.fit(y, floc=10)
pdf_fitted = normal.pdf(x, *param)
plt.plot(x, pdf_fitted, color='r')

# Drawing the histogram
plt.hist(y, normed=True, bins=50)

plt.show()
```

## Graphs from datasets

Source https://www.kaggle.com/unsdsn/world-happiness/data

```python
import pandas as pd
```

```python
happy = pd.read_csv('../../BDINTESA3/Data_Science/data/happiness_2017.csv')
```

```python
happy.columns
```

### Histograms

```python
happy.hist()
plt.show()
```

```python
happy['Freedom'].plot('hist')
plt.title('Freedom')
plt.show()
```

```python
plt.hist(happy['Freedom'], 30, facecolor="green", alpha=0.5)
plt.title('Freedom')
plt.show()
```

### Boxplots

```python
# Health (Life Expectancy): The extent to which Life expectancy contributed to the calculation of the Happiness Score.
happy.boxplot(column='Health..Life.Expectancy.')

plt.title('Health..Life.Expectancy.')

plt.show()
```

```python
happy['Health..Life.Expectancy.'].describe()
```

### Scatter plots

```python
happy.plot(x='Happiness.Score', y='Economy..GDP.per.Capita.', kind='scatter')
plt.title('Scatterplot between happiness rank and GDP')

plt.show()
```

```python
plt.plot(happy['Happiness.Score'], happy['Family'], marker='.', linestyle='', alpha=0.7, c='red')
plt.show()
```

### CDF (Cumulative Distribution Function)

```python
happy['Economy..GDP.per.Capita.'].plot(kind='hist', bins=50, normed=True, cumulative=True)
plt.title('CDF of the GDP contribution to the Happiness Score')

plt.show()
```

```python
happy['Trust..Government.Corruption.'].plot(kind='hist', bins=50, normed=True, cumulative=True)
plt.title('CDF of the Perception of Corruption contribution to the Happiness Score')

plt.show()
```

## Color by a column and alpha parameter

```python
survey = pd.read_csv('../../BDINTESA3/Data_Science/data/WA_American-Time-Use-Survey-lite.csv')
```

```python
plt.scatter(x=survey.Age, y=survey['Food & Drink Prep'], c=survey['Age'], alpha=0.07)

plt.title('Scatter plot on Food & Drink Prep and Age')
plt.xlabel('Age')
plt.ylabel('Food & Drink Prep')

plt.show()
```

## Figure size (and timing execution)

```python
%%time
plt.figure(figsize=(8,6))
plt.scatter(x=survey.Age, y=survey['Weekly Earnings'], c=survey['Age'], alpha=0.04)

plt.title('Scaterplot on Weekly Earnings and Age')
plt.xlabel('Age')
plt.ylabel('Weekly Earnings')

plt.show()
```
