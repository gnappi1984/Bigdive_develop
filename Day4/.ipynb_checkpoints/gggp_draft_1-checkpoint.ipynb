```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
time_survey = pd.read_csv('../../BDINTESA3/Data_Science/data/WA_American-Time-Use-Survey-lite.csv')
```

```python
time_survey_sel = time_survey.loc[~time_survey['Age Range'].isin(['0-19','60-69','70-79', '0-19', '80+']), :].copy()
```

```python
time_survey_sel = time_survey_sel.loc[time_survey_sel['Employment Status']=='Employed', :].copy()
```

```python
time_survey_sel = time_survey_sel.loc[time_survey_sel['Children'] > 0, :].copy()
```

```python
time_survey_sel['Time with Children'] = time_survey_sel['Caring for Children'] + time_survey_sel['Playing with Children']
```

```python
plt.style.use('ggplot')
```

```python
g = sns.catplot(data = time_survey_sel, kind = 'point', x = 'Year',y ='Time with Children', ci = None)
g.set_xticklabels(rotation=45)
g.savefig("./output.png", dpi = 1000)
```

```python
g = sns.catplot(data = time_survey_sel, kind = 'point', x = 'Year',y ='Playing with Children')
g.set_xticklabels(rotation=45)
```

```python
g = sns.catplot(data = time_survey_sel, kind = 'point', x = 'Year',y ='Caring for Children')
g.set_xticklabels(rotation=45)
```

```python
def slice_year(y):
    if y < 2009: 
        return 'A'
    else: 
        return 'B'

time_survey_sel['Periodo'] = time_survey_sel['Year'].apply(lambda per: slice_year(per))
```

```python
dtf_4_stack = (time_survey_sel.loc[:, ['Year','Caring for Children', 'Playing with Children', 'Time with Children']]
 .groupby('Year')
 .mean()
 .reset_index()
 .rename(columns = {'Caring for Children': 'Caring', 'Playing with Children': 'Playing', 'Time with Children':'Total' }))
```

```python
plt.stackplot( dtf_4_stack['Year'],
              [dtf_4_stack['Caring'], dtf_4_stack['Playing']], 
              labels=['Caring','Playing'])

plt.legend()
axes = plt.gca()
axes.set_ylim([40,85])
```

```python
plt.plot( dtf_4_stack['Year'],
               dtf_4_stack['Total'])
```

```python
g = sns.catplot(data = time_survey_sel, hue = 'Gender',kind = 'point', x = 'Age',y ='Time with Children',sharey = True, ci = None)
g.set_xticklabels(rotation=45, step = 5)
```

```python
time_survey_sel_g = time_survey_sel.groupby(['Age Range','Periodo']).mean()
```

```python
time_survey_sel_g.reset_index(inplace=True)
```

```python
time_survey_sel_g['Time with Children']
```

```python
time_survey_sel_g['Age Range'][time_survey_sel_g['Periodo']=='A']
```

```python
N = 4
menMeans = (time_survey_sel_g['Time with Children'][time_survey_sel_g['Periodo']=='A'])
womenMeans = (time_survey_sel_g['Time with Children'][time_survey_sel_g['Periodo']=='B'])
ind = np.arange(N)    # the x locations for the groups
width = 0.35 

p1 = plt.bar(ind, menMeans, width)
p2 = plt.bar(ind+0.1, womenMeans, width,
             bottom=0)

plt.ylabel('Tempo')
plt.title('Tempo speso con i figli')
plt.xticks(ind, (time_survey_sel_g['Age Range'][time_survey_sel_g['Periodo']=='A']))
plt.yticks(np.arange(0, 120, 10))
plt.legend((p1[0], p2[0]), ('Pre 2008', 'Post 2008'))

plt.show()
```
